```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Misalnya df_interpolated sudah kamu dapatkan dari interpolasi sebelumnya
# df_interpolated = df.interpolate(...)

df_encoded = orilatih.copy()

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Iterasi pada setiap kolom bertipe objek (kategorikal)
for col in df_encoded.select_dtypes(include='object').columns:
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))

# Cek hasil
print(df_encoded.head())
```

```python
import pandas as pd

# Asumsikan orilatih sudah didefinisikan sebelumnya
df_interpolated = df_encoded.copy()

# Interpolasi nilai NaN pada kolom numerik
df_interpolated = df_interpolated.interpolate(method='linear', axis=0, limit_direction='forward')

# Cek jumlah nilai NaN per kolom
null_summary = df_interpolated.isnull().sum()

# Cetak hasilnya
if null_summary.sum() == 0:
    print("âœ… Semua nilai NaN berhasil diisi setelah interpolasi.")
else:
    print("âš ï¸ Masih terdapat nilai NaN setelah interpolasi pada kolom berikut:")
    print(null_summary[null_summary > 0])
```

```python
df = df_interpolated.drop(columns=['OWN_CAR_AGE'])
```

```python
df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from imblearn.over_sampling import SMOTE
import pandas as pd

# Asumsikan df sudah di-handle missing values
# dan TARGET adalah nama kolom target

# 1. Identifikasi kolom numerik murni
numeric_cols = df.select_dtypes(include='number').columns.tolist()
numeric_cols.remove('TARGET')  # pastikan kolom target tidak ikut dinormalisasi
numeric_to_scale = numeric_cols.copy()

# 2. Pisahkan fitur dan label
X = df.drop(columns=['TARGET'])
y = df['TARGET']

# 3. Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 4. Normalisasi
scaler = MinMaxScaler()
X_train.loc[:, numeric_to_scale] = scaler.fit_transform(X_train[numeric_to_scale])
X_test.loc[:, numeric_to_scale] = scaler.transform(X_test[numeric_to_scale])

# 5. SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# 6. Cek distribusi
print("Sebelum SMOTE:\n", y_train.value_counts())
print("Sesudah SMOTE:\n", y_train_resampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1. Logistic Regression
logreg = LogisticRegression(max_iter=1000, random_state=42)
logreg.fit(X_train_resampled, y_train_resampled)
y_pred_logreg = logreg.predict(X_test)

print("\n=== Logistic Regression ===")
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
print(confusion_matrix(y_test, y_pred_logreg))
print(classification_report(y_test, y_pred_logreg))

# 2. Decision Tree
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train_resampled, y_train_resampled)
y_pred_dtree = dtree.predict(X_test)

print("\n=== Decision Tree ===")
print("Accuracy:", accuracy_score(y_test, y_pred_dtree))
print(confusion_matrix(y_test, y_pred_dtree))
print(classification_report(y_test, y_pred_dtree))

# 3. Naive Bayes
nb = GaussianNB()
nb.fit(X_train_resampled, y_train_resampled)
y_pred_nb = nb.predict(X_test)

print("\n=== Naive Bayes ===")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print(confusion_matrix(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.tree import DecisionTreeClassifier
import joblib  # untuk ekspor model

# 1. Inisialisasi dan training model Decision Tree
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train_resampled, y_train_resampled)

# 2. Prediksi di test set (opsional evaluasi)
y_pred_dtree = dtree.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Akurasi:", accuracy_score(y_test, y_pred_dtree))
print(confusion_matrix(y_test, y_pred_dtree))
print(classification_report(y_test, y_pred_dtree))

# 3. Ekspor model ke file .pkl
joblib.dump(dtree, 'decision_tree_model.pkl')
print("Model berhasil disimpan ke 'decision_tree_model.pkl'")
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Misalnya df_interpolated sudah kamu dapatkan dari interpolasi sebelumnya
# df_interpolated = df.interpolate(...)

df_encoded_test = data_test.copy()

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Iterasi pada setiap kolom bertipe objek (kategorikal)
for col in df_encoded_test.select_dtypes(include='object').columns:
    df_encoded_test[col] = le.fit_transform(df_encoded_test[col].astype(str))

# Cek hasil
print(df_encoded_test.head())
```

```python
import pandas as pd

# Asumsikan orilatih sudah didefinisikan sebelumnya
df_interpolated_test = df_encoded_test.copy()

# Interpolasi nilai NaN pada kolom numerik
df_interpolated_test = df_interpolated_test.interpolate(method='linear', axis=0, limit_direction='forward')

# Cek jumlah nilai NaN per kolom
null_summary_test = df_interpolated_test.isnull().sum()

# Cetak hasilnya
if null_summary_test.sum() == 0:
    print("âœ… Semua nilai NaN berhasil diisi setelah interpolasi.")
else:
    print("âš ï¸ Masih terdapat nilai NaN setelah interpolasi pada kolom berikut:")
    print(null_summary_test[null_summary_test > 0])
```

```python
df_test = df_interpolated_test.drop(columns=['OWN_CAR_AGE'])
```

```python
# 1. Daftar kolom yang akan diisi null-nya dengan median
median_fill_cols = [
    'YEARS_BUILD_AVG', 'COMMONAREA_AVG', 'ELEVATORS_AVG', 'FLOORSMIN_AVG',
    'LANDAREA_AVG', 'LIVINGAPARTMENTS_AVG', 'NONLIVINGAPARTMENTS_AVG', 'NONLIVINGAREA_AVG',
    'YEARS_BUILD_MODE', 'COMMONAREA_MODE', 'ELEVATORS_MODE', 'FLOORSMIN_MODE',
    'LANDAREA_MODE', 'LIVINGAPARTMENTS_MODE', 'NONLIVINGAPARTMENTS_MODE', 'NONLIVINGAREA_MODE',
    'YEARS_BUILD_MEDI', 'COMMONAREA_MEDI', 'ELEVATORS_MEDI', 'FLOORSMIN_MEDI',
    'LANDAREA_MEDI', 'LIVINGAPARTMENTS_MEDI', 'NONLIVINGAPARTMENTS_MEDI', 'NONLIVINGAREA_MEDI'
]

# 2. Isi nilai kosong dengan median dari masing-masing kolom
for col in median_fill_cols:
    if col in df_test.columns:
        median_value = df_test[col].median()
        df_test[col].fillna(median_value, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# 1. Salin dataframe tes
df_test_normalized = df_test.copy()

# 2. Deteksi kolom yang sebelumnya bertipe objek (kolom yang telah di-label encode)
# Ini harus dilakukan pada data sebelum encoding, tapi kita bisa anggap kamu punya daftarnya dari data latih:
label_encoded_cols = orilatih.select_dtypes(include='object').columns.tolist()

# 3. Identifikasi semua kolom numerik di df_test
numeric_cols = df_test_normalized.select_dtypes(include='number').columns.tolist()

# 4. Tentukan kolom yang akan dinormalisasi (numerik tapi bukan hasil label encoding)
cols_to_scale = [col for col in numeric_cols if col not in label_encoded_cols]

# 5. Normalisasi dengan MinMaxScaler
scaler = MinMaxScaler()
df_test_normalized[cols_to_scale] = scaler.fit_transform(df_test_normalized[cols_to_scale])

# 6. Cek hasil
print(df_test_normalized.head())
```

```python
import joblib

# 1. Muat model dari file pkl
model = joblib.load('decision_tree_model.pkl')

# 2. Lakukan prediksi
# (Pastikan df_test_normalized sudah memiliki kolom-kolom yang sesuai dengan training)
y_pred = model.predict(df_test_normalized)

# 3. Tambahkan hasil prediksi sebagai kolom baru
df_test_normalized['Target'] = y_pred

# 4. (Opsional) Cek hasil
print(df_test_normalized[['Target']].head())
```
